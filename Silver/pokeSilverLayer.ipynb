```python
spark.conf.set("fs.azure.account.auth.type.<storage-account>.dfs.core.windows.net", "OAuth")
spark.conf.set("fs.azure.account.oauth.provider.type.<storage-account>.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set("fs.azure.account.oauth2.client.id.<storage-account>.dfs.core.windows.net", "<application-id>")
spark.conf.set("fs.azure.account.oauth2.client.secret.<storage-account>.dfs.core.windows.net", "service_credential")
spark.conf.set("fs.azure.account.oauth2.client.endpoint.<storage-account>.dfs.core.windows.net", "https://login.microsoftonline.com/<directory-id>/oauth2/token")
```

```python
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("abfss://bronze@dataprojectpikachu.dfs.core.windows.net/pokemon_full_20251125_193829.csv")

display(df)
```

```python
from pyspark.sql import functions
from pyspark.sql.types import *
```

```python
df = df.withColumn("abilities_array", split(col("abilities"), ","))
df = (
    df.withColumn("ability_1", trim(expr("abilities_array[0]")))
      .withColumn("ability_2", trim(expr("get(abilities_array, 1)")))
      .withColumn("ability_3", trim(expr("get(abilities_array, 2)")))
)
df = df.drop( "abilities_array", "abilities")
silver_cleaned_path = "abfss://silver@dataprojectpikachu.dfs.core.windows.net/pokemon_cleaned"
df.write.mode("overwrite").parquet(silver_cleaned_path)

df = df.withColumn("types_array", split(col("types"), ","))
df = df.withColumn("type", explode(col("types_array")))
df = df.withColumn("type", trim(col("type")))
df = df.drop("types_array", "types")
silver_types_path = "abfss://silver@dataprojectpikachu.dfs.core.windows.net/pokemon_types"
df.write.mode("overwrite").parquet(silver_types_path)
```
